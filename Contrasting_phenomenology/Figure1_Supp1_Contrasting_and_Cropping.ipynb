```python
# Figure 1 time 1
# read  tiff using skimage
f1t1 = io.imread('/Users/dylanschaff/Library/CloudStorage/GoogleDrive-dyschaff@sydshafferlab.com/My Drive/White_Shared/cloud/ajf_paper/AJF014/2022_09_13_timepoint1_scans/converted/220913_AJF014_tert_cocl2_full/220913_AJF014_tert_cocl2_full_B3.tif', as_gray=True)
# Rescale the intensity based on contrasting range
f1t1_stretch = exposure.rescale_intensity(f1t1, in_range=(800,1200), out_range='dtype').astype(np.float32)
f1t1_fixed = f1t1_stretch[1073:13286-1073,1075:13312-1075] # crop to remove the extra imaging 
plt.imshow(f1t1_fixed,cmap='gray', vmin=0, vmax=255)
# 1 pixel  = 3.25 uM
scalebar = AnchoredSizeBar(plt.gca().transData,
                           1538.4615384615, '5 mm', 'lower right',
                           pad=0.1,
                           color='red',
                           frameon=False,
                           size_vertical=1)
plt.gca().add_artist(scalebar)
Path("Figure1/Time1").mkdir(parents=True, exist_ok=True)
plt.savefig('Figure1/Time1/contrasted_scalebar_220913_AJF014_tert_cocl2_full_B3.pdf')
plt.show()
# Save the full image 
io.imsave('Figure1/Time1/contrasted_220913_AJF014_tert_cocl2_full_B3.tif', f1t1_fixed)
io.imsave('Figure1/Time1/contrasted_220913_AJF014_tert_cocl2_full_B3.jpg', f1t1_fixed)



```
